```python
from tkinter import *
from tkinter import ttk as ttk
import random
import csv
from flashcard import *
import turtle
import tkinter as tk
#our class does not work unless we wildcard it in

def flip(flashcard_list,the_canvas,index_of_card):
    '''flips card for the user'''
    the_canvas.delete("all")
    current_card=flashcard_list[index_of_card]
    the_canvas.create_rectangle(0,0,500,500,fill="white",outline="")
    
    if(current_card.side_up=="question side"):
        the_canvas.create_rectangle(0,0,500,600,fill="aliceblue",outline="black")
        the_canvas.create_text(270,170,width=530,font="Rockwell",text=current_card.answer)
        current_card.side_up="answer side"
    ##slight bug that you have to click flip twice, however it does change the color and outline
    ##of the answer side
    else:
        the_canvas.create_rectangle(0,0,500,500,fill="white",outline="")
        the_canvas.create_text(270,170,width=530,font="Rockwell",text=current_card.question)
        current_card.side_up="question side"
    global card_index
    card_index=index_of_card
    
def next_card(flashcard_list,the_canvas,index_of_card):
    '''moves to next card, if no new next card, will go to first card'''
    the_canvas.delete("all")
    the_canvas.create_rectangle(0,0,500,500,fill="white",outline = "")
    global card_index
    if(index_of_card+1>=len(flashcard_list)):
        next_card_on_screen=flashcard_list[0]
        card_index=0
    else:
        next_card_on_screen=flashcard_list[index_of_card+1]
        card_index=index_of_card+1
    the_canvas.create_text(270,170,width=530,font="Rockwell",text=next_card_on_screen.question)
    next_card_on_screen.side_up="Q"
    
def previous_card(flashcard_list,the_canvas,index_of_card):
    '''moves to previous card, if no new next card, will go to last card'''
    the_canvas.delete("all")
    the_canvas.create_rectangle(0,0,500,500,fill="white",outline="")
    global card_index
    
    if(index_of_card-1<=-1):
        next_card_on_screen=flashcard_list[len(flashcard_list)-1]
        card_index=len(flashcard_list)-1
    
    else:
        next_card_on_screen=flashcard_list[index_of_card-1]
        card_index=index_of_card-1
    
    the_canvas.create_text(270,170,width=530,font="Rockwell",text=next_card_on_screen.question)
    next_card_on_screen.side_up="Q"

def draw_line(tur, pos1, pos2):
    # print("Drawing from", pos1, "to", pos2)  # Uncomment for tracing the algorithm.
    tur.penup()
    tur.goto(pos1[0], pos1[1])
    tur.pendown()
    tur.goto(pos2[0], pos2[1])
    
def recursive_draw(tur, x, y, width, height, count):
    draw_line(tur,[x + width * 0.25, height // 2 + y],[x + width * 0.75, height // 2 + y],)
    draw_line(tur,[x + width * 0.25, (height * 0.5) // 2 + y],[x + width * 0.25, (height * 1.5) // 2 + y],)
    draw_line(tur,[x + width * 0.75, (height * 0.5) // 2 + y],[x + width * 0.75, (height * 1.5) // 2 + y],)

    if count <= 0:  # The base case
        return
    else:  # The recursive step
        count -= 1
        # Top left
        recursive_draw(tur, x, y, width // 2, height // 2, count)
        # Top right
        recursive_draw(tur, x + width // 2, y, width // 2, height // 2, count)
        # Bottom left
        recursive_draw(tur, x, y + width // 2, width // 2, height // 2, count)
        # Bottom right
        recursive_draw(tur, x + width // 2, y + width // 2, width // 2, height // 2, count)
    

def study_break():
    if __name__ == "__main__":
    
    ####we are really only able to run this once. 
    #### After that, you'd have to close, clear, and restart the program.
    
        screen = turtle.Screen()
        screen.setup(800, 800)
        screen.title("Study Break")
        screen.bgcolor("blue")
    # Turtle artist (pen) setup
        artist = turtle.Turtle()
        artist.hideturtle()
        artist.pensize(5)
        artist.color("lightgreen")
        artist.speed(5)
        #artist.speed(10)
    # Initial call to recursive draw function
        recursive_draw(artist, - 700 / 2, - 700 / 2, 700, 700, 3)
    # Every Python Turtle program needs this (or an equivalent) to work correctly.
        turtle.mainloop()

    
def finalize(flashcard_list):
    '''shuffles cards randomly to be displaced to the user interface'''
    random.shuffle(flashcard_list)
    the_canvas = Canvas(tab2,width=550,height=370, highlightthickness=0)
    the_canvas.create_text(270,170,width=530,font="Consolas",text=flashcard_list[0].question)
    
    global card_index
    card_index=0
    
    the_canvas.configure(bg="white")
    the_canvas.place(x=225,y=60)
    
    flip_button=Button(tab2, text="Flip Card",font=("fixedsys",30),command=lambda:flip(flashcard_list,the_canvas,card_index),bg="black",fg="white")
    flip_button.pack(side=BOTTOM,pady=40)
    next_button=Button(tab2, text="Next",font=("fixedsys",30),command=lambda:next_card(flashcard_list,the_canvas,card_index),bg="#2a19e3",fg="white")
    next_button.place(x=640,y=491)
    prev_button=Button(tab2, text="Prev",font=("fixedsys",30),command=lambda:previous_card(flashcard_list,the_canvas,card_index),bg="#2a19e3",fg="white")
    prev_button.place(x=250,y=491)
    
def get_flashcards(entry,label,button):
    '''entry takes in .csv file and turns it into a list of tuples'''
    try:
        the_file=open(entry.get(),"r")
        entry.delete(0,END)
        the_file.seek(0)
        #print("the file is",the_file)
        file_reader=csv.reader(the_file)
        #print("the file_reader is", file_reader)
        header1=next(file_reader,None)
        flashcard_list=list()
        
        for line in file_reader:
            flashcard_list.append(flashcard(line[0],line[1]))
        #print(flashcard_list)
        
        finalize(flashcard_list)
    except:
        label.configure(text="oops no such file found")
        entry.delete(0,END)
        
def open_help():
    '''gives secondary window (popup) with user instructions and suggestions'''
    secondary_window =Toplevel()
    secondary_window.title("Instructions and Help")
    secondary_window.config(width=700, height=500)
    
    title=ttk.Label(secondary_window, text="*Help for importing file for flashcard use*", font=("Times 20 bold"))
    title.pack()
    
    instruct2 = ttk.Label(secondary_window, text="Import your file by typing in the filename \n -Make sure you download your file first.", font=("Consolas", 15))
    instruct2.pack()
    instruct2a = ttk.Label(secondary_window, text="Make sure your file is in .csv form.", font=("Consolas", 15))
    instruct2a.pack()
    instruct2b = ttk.Label(secondary_window, text="To change your file in Excel, go to Export > Change File Type > .csv > Save.", font=("Consolas", 15))
    instruct2b.pack()
    instruct3a = ttk.Label(secondary_window, text="Go to the Flashcards tab to study your deck.", font=("Consolas", 15))
    instruct3a.pack()
    instruct3b = ttk.Label(secondary_window, text="Sometimes you have to click Flip twice", font=("Consolas", 15))
    instruct3b.pack()
    instruct4a = ttk.Label(secondary_window, text="Clear outputs + Restart Kernel after Study break to get it to run again", font=("Consolas", 15))
    instruct4a.pack()

def main():
    '''this is essentially main, kicks everything off and begins the *game* haha'''
    start_button.destroy()
    start_label.destroy()
    
    flashcard_entry=Entry(tab1,width=80,borderwidth=15)
    flashcard_entry.place(x=245,y=375)
    
    entry_label=Label(tab1, text="Enter .csv file name \n (put .csv at end) ",font=("fixedsys",30),wraplength=590,bg="bisque3",fg="White",justify="center")
    entry_label.place(x=280,y=280)

    entry_button=Button(tab1, text="use flashcard set",font=("fixedsys",30),command=lambda:get_flashcards(flashcard_entry,entry_label,entry_button),bg="#2a19e3",fg="white",padx=73)
    entry_button.place(x=245,y=450)
    
    ###the rest of our data is passed by using the button in Tkinter, which calls a function  or preset module as a command

window=Tk()
window.configure(bg="bisque3")
#window.resizable(False, False)
window.geometry("1001x601")
window.title("Flashcard-App")

tabControl = ttk.Notebook(window)
tab1 = Frame(tabControl)
tab2 = Frame(tabControl)

tabControl.add(tab1, text ='Instructions & File')
tabControl.add(tab2, text ='Flashcards')
tabControl.pack(expand = 1, fill ="both")

help_button= Button(window,text="Help",font=("fixedsys",30),command=open_help,bg="red")
help_button.place(x=0, y=25)
study_button= Button(window,text="Study Break",font=("fixedsys",30),command=study_break,bg="#708090")
study_button.place(x=750, y=25)
start_label=Label(text="FLASHCARDS",font=("fixedsys",155),bg="bisque3",fg="burlywood4",pady=90)
start_label.place(x=245,y=100)
start_button=Button(text="START STUDYING",font=("fixedsys",30),command=main,bg="#2a19e3",fg="white",anchor=CENTER,padx=99)
start_button.place(x=246,y=450)
window.mainloop()
```
